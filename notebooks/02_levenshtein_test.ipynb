```python
import Levenshtein 

def levenshtein_similarity(text1: str, text2: str) -> float:
    distance = Levenshtein.distance(text1, text2)
    max_len = max(len(text1), len(text2))
    return 1 - (distance / max_len)

with open("datasets/doc_original.txt", encoding="utf-8") as f:
    doc1 = f.read()
with open("datasets/doc_paraphrase.txt", encoding="utf-8") as f:
    doc2 = f.read()

score = levenshtein_similarity(doc1, doc2)
print(f"Score Levenshtein : {score:.2%}")
```
