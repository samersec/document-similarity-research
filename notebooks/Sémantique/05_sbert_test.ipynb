```python
# 
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer('paraphrase-multilingual-MiniLM-L12-v2')  # modèle multilingue, adapté au français

with open("../datasets/doc_original.txt", encoding="utf-8") as f:
    doc1 = f.read()
with open("../datasets/doc_paraphrase.txt", encoding="utf-8") as f:
    doc2 = f.read()

embeddings = model.encode([doc1, doc2])
score = util.cos_sim(embeddings[0], embeddings[1]).item()
print(f"Score SBERT (similarité sémantique) : {score:.2%}")
```
