```python
from pathlib import Path


def resolve_data_file(filename: str) -> Path:
    candidates = [
        Path.cwd() / filename,
        Path.cwd() / "datasets" / filename,
        Path.cwd() / "notebooks" / "datasets" / filename,
        Path.cwd().parent / "notebooks" / "datasets" / filename,
    ]
    for candidate in candidates:
        if candidate.exists():
            return candidate
    raise FileNotFoundError(f"Impossible de trouver {filename} dans les emplacements attendus.")


def jaccard_similarity(text1: str, text2: str) -> float:
    # 1. Convertir chaque texte en un ensemble de mots (tokens)
    set1 = set(text1.lower().split())
    set2 = set(text2.lower().split())

    # 2. Calculer l'intersection et l'union des deux ensembles
    intersection = set1.intersection(set2)
    union = set1.union(set2)

    # 3. Retourner le score Jaccard
    return len(intersection) / len(union)


# Chargement des documents
with open(resolve_data_file("doc_original.txt"), encoding="utf-8") as f:
    doc1 = f.read()
with open(resolve_data_file("doc_paraphrase.txt"), encoding="utf-8") as f:
    doc2 = f.read()

# Calcul du score
score = jaccard_similarity(doc1, doc2)
print(f"Score Jaccard : {score:.2%}")

# Bonus : afficher les mots communs pour comprendre le résultat
set1 = set(doc1.lower().split())
set2 = set(doc2.lower().split())
print(f"\nMots communs ({len(set1 & set2)}) : {sorted(set1 & set2)}")
print(f"\nTaille vocabulaire doc1 : {len(set1)}")
print(f"Taille vocabulaire doc2 : {len(set2)}")
```
